```python
# ==============================================================
#  FUZZY SET AND OPERATIONS - EASY & EXPLAINED VERSION
# ==============================================================

import numpy as np
import matplotlib.pyplot as plt

# --------------------------------------------------------------
# 1️⃣ DEFINE THE UNIVERSAL SET (UNIVERSE OF DISCOURSE)
# --------------------------------------------------------------
# The universal set (U) represents all possible values in our domain.
# For example, we are considering temperatures from 0°C to 50°C.
# np.linspace(0, 50, 11) → creates 11 equally spaced points between 0 and 50:
# [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
U = np.linspace(0, 50, 11)

# --------------------------------------------------------------
# 2️⃣ DEFINE TWO FUZZY SETS (A AND B)
# --------------------------------------------------------------
# A fuzzy set assigns each element in U a membership value μ(x) between 0 and 1.
# μ(x) = 0 → not a member of the set
# μ(x) = 1 → completely a member of the set
# Values between 0 and 1 show partial membership.

# Fuzzy Set A represents “Warm” temperatures
# Fuzzy Set B represents “Hot” temperatures
A = np.array([0.0, 0.1, 0.3, 0.5, 0.7, 0.8, 0.9, 1.0, 1.0, 1.0, 1.0])  # Warm
B = np.array([0.0, 0.0, 0.1, 0.2, 0.4, 0.6, 0.8, 0.9, 1.0, 1.0, 1.0])  # Hot

# (Optional Alternative)
# Instead of defining manually, you can generate fuzzy sets using linear functions:
# A = np.clip((5 - U) / 5, 0, 1)   # Decreasing from 1 to 0 as U increases
# B = np.clip((U - 5) / 5, 0, 1)   # Increasing from 0 to 1 as U increases

# --------------------------------------------------------------
# 3️⃣ PERFORM FUZZY OPERATIONS
# --------------------------------------------------------------
# In fuzzy logic, we use mathematical operations to combine or modify fuzzy sets.

# ➤ (i) Union (A ∪ B)
#    - Represents “A or B”
#    - For each element, take the maximum membership value from A or B
union = np.maximum(A, B)

# ➤ (ii) Intersection (A ∩ B)
#    - Represents “A and B”
#    - For each element, take the minimum membership value from A and B
intersection = np.minimum(A, B)

# ➤ (iii) Complement (¬A)
#    - Represents “not A”
#    - For each element, subtract its membership from 1
complement_A = 1 - A

# ➤ (iv) Set Difference (A - B)
#    - Represents “A and not B”
#    - Formula: min(A, 1 - B)
difference = np.minimum(A, 1 - B)

# product = A * B
# algebric_sum = A + B - (A * B)
# bounded_sum = np.minimum(1, A + B)
# bounded_difference = np.maximum(0, A - B)

# --------------------------------------------------------------
# 4️⃣ PRINT RESULTS (NUMERICAL OUTPUT)
# --------------------------------------------------------------
print("===== FUZZY OPERATION RESULTS =====\n")
print("Universal Set (U):", U)
print("Set A (Warm):", A)
print("Set B (Hot):", B)
print("\nUnion (A ∪ B):", union)
print("Intersection (A ∩ B):", intersection)
print("Complement of A (¬A):", complement_A)
print("Set Difference (A - B):", difference)

# --------------------------------------------------------------
# 5️⃣ PLOT ALL FUZZY SETS AND THEIR OPERATIONS
# --------------------------------------------------------------
# plt.figure(figsize=(8, 5))

# Original fuzzy sets
plt.plot(U, A, 'b-', linewidth=2, label='Set A (Warm)')
plt.plot(U, B, 'r-', linewidth=2, label='Set B (Hot)')

# Fuzzy operations (with dashed styles for distinction)
plt.plot(U, union, 'g--', linewidth=2, label='Union (A ∪ B)')
plt.plot(U, intersection, 'm--', linewidth=2, label='Intersection (A ∩ B)')
plt.plot(U, complement_A, 'c--', linewidth=2, label='Complement (¬A)')
plt.plot(U, difference, 'y-.', linewidth=2, label='Set Difference (A - B)')

# Add chart labels, title, legend, and grid
plt.xlabel('Temperature (°C)')
plt.ylabel('Membership Value (μ)')
plt.title('Fuzzy Set Operations Visualization')
plt.legend()
plt.grid(True)
# plt.ylim(-0.05, 1.05)
plt.show()
```
