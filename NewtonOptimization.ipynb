```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to optimize
def f(x):
    return x**4 - 3*x**3 + 2

# First derivative
def f_prime(x):
    return 4*x**3 - 9*x**2

# Second derivative
def f_double_prime(x):
    return 12*x**2 - 18*x

# Newton's method for optimization
def newton_optimization(x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        f1 = f_prime(x)
        f2 = f_double_prime(x)
        if f2 == 0:
            print("Second derivative zero. Stopping iteration.")
            break
        x_new = x - f1 / f2
        if abs(x_new - x) < tol:
            print(f"Converged in {i+1} iterations.")
            return x_new
        x = x_new
    print("Max iterations reached.")
    return x

# Initial guess
x0 = 2.0
opt_x = newton_optimization(x0)
print(f"Optimal x: {opt_x:.6f}")
print(f"Function value at optimum f(x): {f(opt_x):.6f}")

x_vals = np.linspace(0, 3, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='f(x)')
plt.scatter(opt_x, f(opt_x), color='red', label=f'Optimal x={opt_x:.2f}')
plt.title("Newton's Method Optimization")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
