```python
# ==========================================================
# Principal Component Analysis (PCA) on the Iris Dataset
# ==========================================================
# Goal: Reduce 4D data (4 features) into 2D for visualization
# Steps:
#   1. Load dataset from CSV file
#   2. Select features (X) and target (y)
#   3. Scale features using StandardScaler
#   4. Apply PCA (reduce to 2 principal components)
#   5. Visualize results in a 2D scatter plot
# ==========================================================

# Import necessary libraries
import pandas as pd                     # For handling tabular data
from sklearn.preprocessing import StandardScaler   # For feature scaling
from sklearn.decomposition import PCA              # For dimensionality reduction
import matplotlib.pyplot as plt          # For plotting graphs

# ----------------------------------------------------------
# 1 Load the dataset
# ----------------------------------------------------------
# Read the iris dataset from a local CSV file
# Make sure 'iris_dataset.csv' is in your working directory
# or replace the file name with its full path
df = pd.read_csv('iris_dataset.csv')

# Display first few rows to confirm successful loading
# print(df.head())

# ----------------------------------------------------------
# 2 Select Features (X) and Target (y)
# ----------------------------------------------------------
# X  Input features used for analysis
# y  Target class (0, 1, 2) representing iris flower types
X = df[['sepal length (cm)', 'sepal width (cm)', 
        'petal length (cm)', 'petal width (cm)']]  # independent variables
y = df['target']  # dependent variable (class label)

# ----------------------------------------------------------
# 3 Feature Scaling
# ----------------------------------------------------------
# PCA is sensitive to feature scales (large values dominate variance)
# StandardScaler standardizes features:
#   - mean = 0
#   - standard deviation = 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# X_scaled is now a NumPy array with scaled values
# Example shape: (150 samples, 4 features)
# print(X_scaled[:5])

# ----------------------------------------------------------
# 4 Apply PCA
# ----------------------------------------------------------
# PCA reduces data dimensions while preserving most variance
# Here, we reduce 4 features  2 principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# X_pca now contains 2 new columns:
#   - PC1: first principal component (most variance)
#   - PC2: second principal component (next most variance)
# print(X_pca[:5])

# ==========================================================
#  Principal Component Analysis (PCA) on the Iris Dataset
# ==========================================================
# Goal: Reduce 4D data (4 features) into 2D for visualization
# Steps:
#   1. Load dataset from CSV file
#   2. Select features (X) and target (y)
#   3. Scale features using StandardScaler
#   4. Apply PCA (reduce to 2 principal components)
#   5. Visualize results in a 2D scatter plot
# ==========================================================

# Import necessary libraries
import pandas as pd                     # For handling tabular data
from sklearn.preprocessing import StandardScaler   # For feature scaling
from sklearn.decomposition import PCA              # For dimensionality reduction
import matplotlib.pyplot as plt          # For plotting graphs

# ----------------------------------------------------------
# 1 Load the dataset
# ----------------------------------------------------------
# Read the iris dataset from a local CSV file
# Make sure 'iris_dataset.csv' is in your working directory
# or replace the file name with its full path
df = pd.read_csv('iris_dataset.csv')

# Display first few rows to confirm successful loading
# print(df.head())

# ----------------------------------------------------------
# 2 Select Features (X) and Target (y)
# ----------------------------------------------------------
# X  Input features used for analysis
# y  Target class (0, 1, 2) representing iris flower types
X = df[['sepal length (cm)', 'sepal width (cm)', 
        'petal length (cm)', 'petal width (cm)']]  # independent variables
y = df['target']  # dependent variable (class label)

# ----------------------------------------------------------
# 3 Feature Scaling
# ----------------------------------------------------------
# PCA is sensitive to feature scales (large values dominate variance)
# StandardScaler standardizes features:
#   - mean = 0
#   - standard deviation = 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# X_scaled is now a NumPy array with scaled values
# Example shape: (150 samples, 4 features)
# print(X_scaled[:5])

# 4. Print Explained Variance
# This shows how much information (variance) is captured by each component.
print("Explained variance ratio by PC1 and PC2:", pca.explained_variance_ratio_)
print("Total explained variance:", sum(pca.explained_variance_ratio_))

# ----------------------------------------------------------
# 4 Apply PCA
# ----------------------------------------------------------
# PCA reduces data dimensions while preserving most variance
# Here, we reduce 4 features  2 principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# X_pca now contains 2 new columns:
#   - PC1: first principal component (most variance)
#   - PC2: second principal component (next most variance)
# print(X_pca[:5])

# ----------------------------------------------------------
# 5 Visualize the PCA-transformed data
# ----------------------------------------------------------
# We'll plot the first two principal components in a 2D scatter plot
plt.figure(figsize=(8, 6))              # Set figure size
plt.scatter(X_pca[:, 0],                # PC1 on x-axis
            X_pca[:, 1],                # PC2 on y-axis
            c=y,                        # Color by target class
            cmap='viridis',             # Use colormap (green  purple)
            edgecolors='k',             # Black edge for points
            s=100)                      # Size of points

# Label the axes and add title
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('PCA on Iris Dataset')

# Display the plot
plt.show()

# ----------------------------------------------------------
#  What youll see:
# ----------------------------------------------------------
# - Each point = one flower sample
# - Color = flower species (setosa, versicolor, virginica)
# - PCA compresses the 4D feature space into 2D
# - Flowers of the same type cluster together
```
